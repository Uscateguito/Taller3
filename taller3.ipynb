```python
# Es necesario instalar pandas, numpy, openpyxl y scipy. Es posible hacer esto con pip.

import pandas as pd
import math
import numpy as np
from pulp import *
import xlsxwriter
import random
from openpyxl import load_workbook
import scipy.stats as st
```

```python
# Constantes

lambdaa = 0.05
miu = 0.08
acumuladordetiempos = 0

# Como no estábamos seguros de si el límite debía ser por algún motivo específico, pensamos en una cola real y pensamos que el
# límite más intuitivo que conocemos es el tiempo que pueden durar atendiendo un servicio, en este sentido, la variable que limita
# la ejecución de el algoritmo, es la columna hora, así:

limitedetiempodeatencion = 200

# Cantidad de servicios que serán atendidos en la cola

serviciosxAtender = 5

# Entidades (Columnas)

evento = []
evento.append("inicio")
# print(Evento)

hora = []
hora.append(0)

# print(hora)

# Numero del servicio que hay que prestar en la cola (se deriva de servicio atendido)
# Como la columna cola es una lista, entonces hay otra variable que representa esta lista
cola = []
elementosenCola = []
cola.append("")


# Tamaño de la cola

tamanioCola = []
tamanioCola.append(len(elementosenCola))

# Numero de confirmación de un servicio terminado

servicioAtendido = 0
servicioAtendidoActualmente = []
servicioAtendidoActualmente.append(servicioAtendido)

# Tiempo restante de llegada = tiempo que falta para que llegue el siguiente cliente

tiempoRestanteLlegada = []
tiempoRestanteLlegada.append((-1/lambdaa)*math.log(random.random()))

# Tiempo restante de servicio = tiempo que falta para terminar de atender al cliente

tiempoRestanteServicio = []
tiempoRestanteServicio.append(0)

# Funcion

# Creamos el caso base

evento.append("llegada")
hora.append(tiempoRestanteLlegada[0])
cola.append("")
tamanioCola.append(len(elementosenCola))
servicioAtendido = 1
servicioAtendidoActualmente.append(servicioAtendido)
tiempoRestanteLlegada.append((-1/lambdaa)*math.log(random.random()))
tiempoRestanteServicio.append((-1/miu)*math.log(random.random()))

# La i, sirve para indicar a partir de cual elemento vamos a empezar a trabajar dado el caso base que es la primera llegada
# la hora acumulada es simplemente una bandera para acabar con la ejecución del algoritmo

i = 2
j= 0
horaAcumulada = 0
strcola = ""

while (horaAcumulada < limitedetiempodeatencion):
    # Si es más demorada la atención que la llegada del siguiente cliente, el cliente debe esperar y se forma una cola
    
    if tiempoRestanteServicio[i-1] > tiempoRestanteLlegada[i-1] or tiempoRestanteServicio[i-1] == 0:
        evento.append("llegada")
        hora.append(hora[i-1] + tiempoRestanteLlegada[i-1])

        if len(elementosenCola) == 0:
            elementosenCola.append(servicioAtendido+1)
            strcola = str(elementosenCola)
            cola.append(strcola)
            j += 1
            
        elif len(elementosenCola) > 0:
            elementosenCola.append(elementosenCola[j-1]+1)
            strcola = str(elementosenCola)
            cola.append(strcola)          
            j += 1
        tamanioCola.append(len(elementosenCola))
        servicioAtendidoActualmente.append(servicioAtendido)
        tiempoRestanteLlegada.append((-1/lambdaa)*math.log(random.random()))
        if(tiempoRestanteServicio[i-1] == 0):
            tiempoRestanteServicio.append((-1/miu)*math.log(random.random()))
        else:
            tiempoRestanteServicio.append(
            tiempoRestanteServicio[i-1] - tiempoRestanteLlegada[i-1])
       

        

    # Si el cliente fue atendido rápido y sale de la cola antes de que llegue el un nuevo cliente
    elif tiempoRestanteServicio[i-1] < tiempoRestanteLlegada[i-1]:
        evento.append("salida")
        hora.append(hora[i-1] + tiempoRestanteServicio[i-1])
        
        if len(elementosenCola) > 0:
            elementosenCola.pop(0)
            j -= 1
            strcola = str(elementosenCola)
            cola.append(strcola)
            tamanioCola.append(len(elementosenCola))
            servicioAtendido += 1
            servicioAtendidoActualmente.append(servicioAtendido)
            tiempoRestanteLlegada.append(tiempoRestanteLlegada[i-1] - tiempoRestanteServicio[i-1])
            tiempoRestanteServicio.append((-1/miu)*math.log(random.random()))
        else:
            strcola = str(elementosenCola)
            cola.append(strcola)
            tamanioCola.append(len(elementosenCola))
            servicioAtendido += 1
            servicioAtendidoActualmente.append(servicioAtendido)
            tiempoRestanteLlegada.append(tiempoRestanteLlegada[i-1] - tiempoRestanteServicio[i-1])
            tiempoRestanteServicio.append(0)

    # print(i)   
    # print(cola)
    horaAcumulada = hora[i]
    print(hora[i])
    i += 1
    

print(evento)      
print(hora)
print(cola)
print(tamanioCola)
print(servicioAtendidoActualmente)
print(tiempoRestanteLlegada)
print(tiempoRestanteServicio)

matriz = np.column_stack((evento, hora, cola, tamanioCola, servicioAtendidoActualmente, tiempoRestanteLlegada, tiempoRestanteServicio))
```

```python
writer=pd.ExcelWriter('colasServidor.xlsx', engine='xlsxwriter')
workbook=writer.book

dataf=pd.DataFrame([matriz[i,:] for i in range (len(evento))], columns=['evento',  'hora', 'cola', 'tamaño Cola', 'servicio Atendido Actualmente', 'tiempo Restante Llegada', 'tiempo Restante Servicio'])

dataf.to_excel(
    writer,
    index=False,
    sheet_name= 'prueba'
)


writer.save()
```
